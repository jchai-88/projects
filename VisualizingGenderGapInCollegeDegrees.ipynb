The Department of Education Statistics(https://nces.ed.gov/programs/digest/2013menu_tables.asp) releases a data set annually containing the percentage of bachelor's degrees granted to women from 1970 to 2012. The data set is broken up into 17 categories of degrees, with each column as a separate category.

Randal Olson, a data scientist at University of Pennsylvania, has cleaned the data set and made it available on his personal website. You can download the dataset Randal compiled here http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv. Here's a preview of the first few rows:

Randal compiled this data set to explore the gender gap in STEM fields, which stands for science, technology, engineering, and mathematics. This gap is reported on often in the news and not everyone agrees that there is a gap.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Initialise variables
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
rbg_dark_blue = (0/255,107/255,164/255)
rbg_orange = (255/255,128/255,14/255)
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']

# Create figure
fig = plt.figure(figsize=(18, 25))


# Create subplots for stem_cats
ind = 1

for i in range (0,6):
    ax = fig.add_subplot(6,3,ind)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[i]], c = rbg_dark_blue, label = 'Women', linewidth = 3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[i]], c = rbg_orange, label = 'Men', linewidth = 3)
    ax.spines["bottom"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[i])
    ax.tick_params(bottom = False, top = False, left = False, right = False, labelbottom = False)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ind += 3
    
    # For toptmost subplot and bottommost subplot, add text annotations
    if i == 0:
        ax.text(2005, 81, 'Women')
        ax.text(2005, 15, 'Men')
    elif i == 5:
        ax.text(2006, 88, 'Men')
        ax.text(2006, 9, 'Women')
        ax.tick_params(labelbottom = True)
        
# Create subplots for lib_arts_cats
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']

ind = 2

for i in range (0,5):
    ax = fig.add_subplot(6,3,ind)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[i]], c = rbg_dark_blue, label = 'Women', linewidth = 3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[i]], c = rbg_orange, label = 'Men', linewidth = 3)
    ax.spines["bottom"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[i])
    ax.tick_params(bottom = False, top = False, left = False, right = False, labelbottom = False)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ind += 3
    
    # For toptmost subplot and bottommost subplot, add text annotations
    if i == 0:
        ax.text(2005, 21, 'Men')
        ax.text(2005, 75, 'Women')
    elif i == 4:
        ax.tick_params(labelbottom = True)

# Create subplots for other_cats
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

ind = 3

for i in range (0,6):
    ax = fig.add_subplot(6,3,ind)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[i]], c = rbg_dark_blue, label = 'Women', linewidth = 3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[i]], c = rbg_orange, label = 'Men', linewidth = 3)
    ax.spines["bottom"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[i])
    ax.tick_params(bottom = False, top = False, left = False, right = False, labelbottom = False)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ind += 3
    
    # For toptmost subplot and bottommost subplot, add text annotations
    if i == 0:
        ax.text(2007, 8, 'Men')
        ax.text(2005, 90, 'Women')
    elif i == 5:
        ax.text(2006, 62, 'Men')
        ax.text(2006, 35, 'Women')
        ax.tick_params(labelbottom = True)

plt.savefig('gender_degrees.png')
plt.show()
```
